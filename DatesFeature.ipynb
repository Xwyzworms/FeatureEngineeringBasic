Dates And Times --> Special type of categorical variable ,instead taking traditional labels { Colors , city , quality } ,it take dates as values -->  Month,day,Year 

 Date variables usually contain a huge number of individual categories, which will expand the feature space dramatically
- Date variables allow us to capture much more information from the dataset if preprocessed in the right way




```
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/DataSet/Index2018.csv")
df.head()
```

```
df["date_dt"] = pd.to_datetime(df["date"])
df.drop("date",inplace=True,axis=1)
df.head()
```

This Stuff will help you out ---> {Extract the year , month ,and day }

```
df["Year"] = df["date_dt"].dt.year
df["Month"] = df["date_dt"].dt.month
df["Day"] = df["date_dt"].dt.day
df.head()
```

Also you Can add new features , like boolean value --> is_quarter , is_endof_year, And also remember if you deal with missing value on date variable ,you need to fill it using its next/before day.

```
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
df.groupby(["Month","Year"])["spx"].sum().plot()
plt.show()
```

What we see here is that SPX is not Unstable,near every month the price drop quitely high but then the price is raised up again
